```python
import pandas as pd
df=pd.read_csv("mymoviedb.csv")
df
```

```python
df.columns.tolist()
```

```python
df['Release_Date'] = pd.to_datetime(df['Release_Date'], errors='coerce')
df['year'] = df['Release_Date'].dt.year

df_2019 = df[df['year'] == 2019]
print(df_2019.shape)
df_2019.head()
```

```python
genre_counts = (
    df_2019['Genre']
    .str.split(',')        # break "Action, Comedy" into ["Action", " Comedy"]
    .explode()             
    .str.strip()         
    .value_counts()
    .head(10)
)
genre_counts
```

```python
top_movies = (
    df_2019[['Title', 'Popularity']]
    .sort_values('Popularity', ascending=False)
    .head(10)
    .set_index('Title')
)
top_movies
```

```python
scatter_data = (
    df_2019[['Title', 'Popularity', 'Vote_Average', 'Vote_Count']]
    .dropna(subset=['Popularity', 'Vote_Average'])
)
print(scatter_data.shape)
scatter_data.to_csv("scatter_2019.csv", index=False)
```
